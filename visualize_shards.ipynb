```python
import tarfile
from PIL import Image
import io
import matplotlib.pyplot as plt

# Path to your .tar file
tar_path = "webdataset_test/shard-00001.tar"

# Open the tar file
with tarfile.open(tar_path, "r") as tar:
    # List all files in the archive
    all_files = tar.getnames()
    print("All files found:", all_files)
    
    # List all PNG files in the archive
    png_files = [m for m in all_files if m.endswith('.png')]
    print("PNG files found:", png_files)
    
    # Display all PNG images
    for png_file in png_files:
        member = tar.getmember(png_file)
        fileobj = tar.extractfile(member)
        img = Image.open(io.BytesIO(fileobj.read()))
        plt.figure()
        plt.title(png_file)
        plt.imshow(img)
        plt.axis('off')
        plt.show()
```
